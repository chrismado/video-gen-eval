# video-gen-eval quick walkthrough

This notebook creates a tiny synthetic video, runs the unified pipeline, and inspects the resulting physics report.

```python
from pathlib import Path

import cv2
import numpy as np

from pipeline.unified_pipeline import UnifiedPipeline
```

```python
artifacts = Path('examples/notebook_artifacts')
artifacts.mkdir(parents=True, exist_ok=True)
video_path = artifacts / 'notebook_demo.avi'
writer = cv2.VideoWriter(str(video_path), cv2.VideoWriter_fourcc(*'MJPG'), 8.0, (96, 96))
for frame_idx in range(10):
    frame = np.zeros((96, 96, 3), dtype=np.uint8)
    frame[18:42, 12 + frame_idx:36 + frame_idx] = (60, 210, 180)
    writer.write(frame)
writer.release()
video_path
```

```python
pipeline = UnifiedPipeline(enable_vbench=False, enable_ivebench=False, enable_tivibench=False, enable_physion=True)
report = pipeline.run(str(video_path))
report.to_dict()
```
